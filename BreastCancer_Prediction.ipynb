```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# loading the dataset
breast_cancer=load_breast_cancer()
breast_cancer
```

```python
# converting the dictionar of data into dataframe using pandas
df_data=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
df_data.head(9)
```

```python
# converting the target into data frame
df_target=pd.DataFrame(breast_cancer.target,columns=['label'])
```

```python
df_data.info()
```

```python
# describing the data to see the their  summary
df_data.describe()
```

```python
# checking the null values
df_data.isnull().sum()
```

```python
#  creating dependent and independent data
x=df_data
y=df_target
```

Splitting of data

```python
# splitting the data for model training into test and train. standardizing and fitting the data for better accuracy
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x_train,x_test,y_train,y_teat=train_test_split(x,y,test_size=0.2,random_state=44)
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

Model training

```python
# importing the essential libraries for creating the neural network. 
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow.keras.layers import Dense, Flatten #type: ignore
from tensorflow.keras.models import Sequential # type: ignore
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint # type: ignore
```

```python
# creating the model of neural network, using flatten--> the input data are, in 2D form. So, flatten help to convert them into 1D.
model=Sequential([
    Flatten(input_shape=((x_train_std.shape[1],))),
    Dense(20,activation='relu'),
    Dense(2,activation='sigmoid')
])
```

```python
# compiling the model, using optimizer, loss and metrics
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# adding callbacks to preventing from overfitting
early_stopping=EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
check_points=ModelCheckpoint('best_mode.h5',monitor='val_loss',save_best_only=True,mode='min')
```

```python
# finally fiti=ting the training data into created NN model with epochs and callbacks
history=model.fit(x_train_std,y_train,validation_split=0.1,epochs=10,callbacks=[early_stopping,check_points])
```

Graphical representation of Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.xlabel('epoches')
plt.xlabel('accuracy')
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoches')
plt.xlabel('losses')
plt.legend(['training loss','validation loss'],loc='upper right')
```

```python
loss, accuracy= model.evaluate(x_test_std,y_teat)
print(accuracy)
```
